```python
from datetime import datetime
import pandas as pd
```

```python
df = pd.read_csv('cotacao_itsa3.csv')
```

```python
#transforma os indices em colunas
df2 = df.reset_index()
```

```python
#Acrescentando novas colunas
df2
```

```python
#seleciona as colunas
df3 = df2[['ticker','data','hora','abertura','fechamento','menor','maior','volume']]
```

```python
#Renomeia colunas
df3
```

```python
import matplotlib.pyplot as plt
```

```python
#evolução diaria

x=df3["data"]
y=df3["fechamento"]
plt.figure(figsize=(10, 5))
plt.plot(x, y, color = 'blue')
plt.xticks(rotation=90)
plt.legend()
plt.xlabel('Dias')
plt.ylabel('Valor')
plt.show()
```
